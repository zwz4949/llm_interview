```python
import torch
import torch.nn as nn
class RMSNorm(nn.Module):
    def __init__(self, dim, eps=1e-6):
        super().__init__()
        self.eps = eps
        # 可学习的缩放参数
        self.weight = nn.Parameter(torch.ones(dim))

    def _norm(self, x):
        # RMSNorm核心操作
        # 不像LayerNorm，RMSNorm没有减去均值的步骤
        return x * torch.rsqrt(x.pow(2).mean(-1, keepdim=True) + self.eps)

    def forward(self, x):
        # 应用可学习的权重
        return self.weight * self._norm(x)
```

## rmsnorm(x) = x/(1/n)*sqrt(x^2)

```python
import torch
import torch.nn as nn
class RMSNorm(nn.Module):
    def __init__(self, dim, eps = 1e-6):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(dim))
        self.eps = eps
    def _norm(self, hidden_states):
        
        hidden_states = hidden_states.to(torch.float32)
        hidden_states = hidden_states * torch.rsqrt(hidden_states.pow(2).mean(-1,keepdim = True) * self.eps)
        return hidden_states
    def forward(self, hidden_states):
        input_dtype = hidden_states.dtype
        hidden_states = self._norm(hidden_states)
        hidden_states = self.weight * hidden_states
        return hidden_states.to(input_dtype)
normlayer = RMSNorm(1024)
normlayer(torch.randn(2,2,1024))
```
