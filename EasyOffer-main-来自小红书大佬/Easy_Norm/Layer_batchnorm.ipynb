```python
import torch

x = torch.tensor([
    [  # 第一个样本
        [1.0, 2.0, 3.0],    # 时间步1
        [4.0, 5.0, 6.0],    # 时间步2
    ],
    [  # 第二个样本
        [7.0, 8.0, 9.0],
        [10.0, 11.0, 12.0],
    ]
])
x.shape
```

```python
import torch.nn as nn

layer_norm = nn.LayerNorm(normalized_shape=3)  # 对最后一个维度进行归一化
output_ln = layer_norm(x)
output_ln
```

```python
batch_norm = nn.BatchNorm1d(num_features=3)  # 对特征维度进行归一化

# 调整张量形状以匹配 BatchNorm1d 的输入要求
x_reshaped = x.permute(0, 2, 1)  # 形状变为 (batch_size, d_model, seq_len)
output_bn = batch_norm(x_reshaped)

# 如果需要恢复原始形状
output_bn = output_bn.permute(0, 2, 1)  # 形状变回 (batch_size, seq_len, d_model)
output_bn
```
